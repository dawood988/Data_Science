# Defining x and y variable

```python
x_train = train_data1.drop(columns=['Salary']).values
y_train = train_data1['Salary'].values
```

```python
x_test = test_data1.drop(columns=['Salary']).values
y_test = test_data1['Salary']
```

# SVM using grid cv

```python
clf = SVC()
```

```python
param_grid = [{'kernel':['rbf'],'gamma':[50,10,5,0.5,0.2],'C':[15,14,13,12,11,10,5,0.1,0.01,0.001]}]
grid_cv = GridSearchCV(clf,param_grid=param_grid,cv = 10)
grid_cv.fit(x_train,y_train)
```

```python
grid_cv.best_params_,grid_cv.best_score_
```

```python
clf = SVC(C=5,gamma=0.2)
clf.fit(x_test,y_test)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test,y_pred)
print('Accuracy =',acc)
confusion_matrix(y_test,y_pred)
```
