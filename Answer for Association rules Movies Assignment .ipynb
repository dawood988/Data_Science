```python
import pandas as pd
import numpy  as np
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
from mlxtend import plotting
```

# Importing the Data sets

```python
df = pd.read_csv(r"C:\Excelr assignments\Assignment 8 Association rules\my_movies.csv")
```

```python
df
```

```python
df1 = df.drop(columns=['V1','V2','V3','V4','V5'])
```

```python
df1
```

# Apriori algorithm

# For support = 0.1 or 10% and confidence = 0.1

```python
Movies = apriori(df1,min_support=0.1,use_colnames=True)
```

```python
Movies
```

```python
rules = association_rules(Movies,metric='lift',min_threshold=0.1)
```

```python
rules
```

```python
rules.sort_values('lift',ascending=False)[0:20]
```

```python
rules[rules.lift>1]
```

# Visualisation of data 

```python
import seaborn as sns
```

```python
sns.pairplot(rules)
```
