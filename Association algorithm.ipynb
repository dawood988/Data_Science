```python
without_duplicates = df.drop_duplicates(['User_ID','Book_Title'],keep='first')
```

```python
without_duplicates
```

```python
user_df = without_duplicates.pivot(index= 'User_ID',
                  columns= 'Book_Title',
                  values= 'Book_Rating',).reset_index(drop=True)
```

```python
user_df
```

```python
user_df.index = without_duplicates.User_ID.unique()
```

```python
user_df
```

# Impute those NaN's with 0 values

```python
user_df.fillna(0,inplace=True)
```

```python
user_df
```

# Calculating the cosine similarities between them

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim = 1 - pairwise_distances(user_df.values,metric='cosine')
```

```python
user_sim
```

# Store the result in the data frame

```python
user_sim_df = pd.DataFrame(user_sim)
```
