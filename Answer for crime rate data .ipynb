# Performing normalization on the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler( with_mean=True, with_std=True)
x = scaler.fit_transform(array)
```

# Applying heirarchical Clustering

```python
dendrogram = hierarchy.dendrogram(hierarchy.linkage(df2,method='single'))
```

```python
dendrogram
```

# Creating the Heirarchical clusters

```python
hc = AgglomerativeClustering(n_clusters=5, metric='euclidean', linkage='single')
```

```python
y_hc = hc.fit_predict(x)
```

```python
H_clusteriod = pd.DataFrame(y_hc,columns=['Heirarchical_cluster'])
```

```python
df = pd.concat([df,H_clusteriod],axis=1)
```

```python
df
```

# K-mean Clustering

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range (1,10):
    kmeans=KMeans(n_clusters=i, random_state=20,n_init=10,)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,10),wcss)
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

# Build a cluster algorithm

```python
KMeans(algorithm=x)
```

```python
cluster_kmeans = KMeans(n_clusters=5,n_init=10,random_state=20)
cluster_kmeans.fit(x)
```

```python
cluster_kmeans.labels_
```

```python
clusteriod_kmeans = pd.DataFrame(cluster_kmeans.labels_,columns=['Clusteriod_Kmeans']) 
```

```python
df = pd.concat([df,clusteriod_kmeans],axis=1)
```
